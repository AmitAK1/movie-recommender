```python
def toListscast(row):
    row = ast.literal_eval(row)
    li=[]
    count=0;
    for i in row:
        if count<3:
            li.append(i['name'])
            count+=1
        else:
            break
    return li
    
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
# to combine words with same meaning(#stemming) ie love loving loves => love
```

```python
def stem(text):
    ls=[]
    for i in text.split():
        ls.append(ps.stem(i))
    return " ".join(ls)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
def recommend(movie):
    li=[]
    movie_index=new_df[new_df['title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    #movie_list 2d matrix h 1st pe index 2nd pe dist
    for i in movie_list:
        #traversing through each row
        li.append(i[0])
        #appending index only not distance calculated
    recommended(li)
    
```

```python
def recommended(index_acq):
    for i in index_acq:
        print(new_df['title'].iloc[i])
```
